```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
%matplotlib inline
```

```python
from IPython.html.widgets import interact
import IPython.html.widgets as widgets
```

```python
data = pd.read_csv('../output/condition_summary.csv')
```

```python
conditions = data[data['Function'] == 'mean'].drop(['Function','Well Name'],axis=1).set_index('Condition').apply(lambda x: x.values/np.linalg.norm(x.values)).index.values
features = data[data['Function'] == 'mean'].drop(['Function','Well Name'],axis=1).set_index('Condition').apply(lambda x: x.values/np.linalg.norm(x.values)).columns
```

```python
normed_data = data[data['Function'] == 'mean'].drop(['Function','Well Name'],axis=1).set_index('Condition').apply(lambda x: x.values/np.linalg.norm(x.values)).values.T
```

```python
plt.figure(figsize=(17,2.5))
plt.imshow(normed_data,interpolation='nearest',cmap='Blues',aspect='auto');
plt.axis('off');
```

```python
from matplotlib._cm import datad
cnames = list(cm.six.iterkeys(datad))
```

```python
def plot_grid(x,y):
    plt.figure(figsize=(x,y))
    plt.imshow(normed_data,interpolation='nearest',cmap='Blues',aspect='auto');
```

```python
fig = plt.figure(figsize=(17,5))
plt.subplots_adjust(hspace=0)

plt.subplot(2,1,1)
plt.imshow(normed_data,interpolation='nearest',cmap='Blues',aspect='auto');
plt.axis('off');

plt.subplot(2,1,2)
plt.imshow(normed_data,interpolation='nearest',cmap='Reds',aspect='auto');
plt.axis('off');

```

```python
from scipy.cluster.hierarchy import *
link = linkage(normed_data.T)
dendrogram(link);
```

```python
new_order = leaves_list(link)
```

```python
reordered_data = normed_data[:,new_order]
```

```python
fig = plt.figure(figsize=(17,10))
plt.subplots_adjust(hspace=0)

plt.subplot(2,1,1)
dendrogram(link);
plt.axis('off');

plt.subplot(2,1,2)
plt.imshow(reordered_data,interpolation='nearest',cmap='Blues',aspect='auto');
plt.axis('off');
```

```python

feature_link = linkage(normed_data)
new_feature_order = leaves_list(feature_link)
doubly_reordered_data = normed_data[new_feature_order,:][:,new_order]

fig = plt.figure(figsize=(17,10))
plt.subplots_adjust(hspace=0,wspace=0)

plt.subplot(2,2,1)
dendrogram(link);
plt.axis('off');

plt.subplot(2,2,3)
plt.imshow(doubly_reordered_data,interpolation='nearest',cmap='Blues',aspect='auto');
plt.axis('off');

plt.subplot(2,2,4)
linkage(normed_data)
dendrogram(feature_link,orientation='left');
plt.axis('off');
```

```python
def plot_new_thing(width, height,height_ratio):
    feature_link = linkage(normed_data)
    new_feature_order = leaves_list(feature_link)
    doubly_reordered_data = normed_data[new_feature_order,:][:,new_order]

    import matplotlib.gridspec as gridspec

    fig = plt.figure(figsize=(width,height))
    plt.subplots_adjust(hspace=0,wspace=0)

    gs = gridspec.GridSpec(2, 2, width_ratios=[5, 1],height_ratios=[1,height_ratio]) 

    plt.subplot(gs[0])
    dendrogram(link);
    plt.axis('off');

    plt.subplot(gs[2])
    plt.imshow(doubly_reordered_data,interpolation='nearest',cmap='Blues',aspect='auto');
    plt.gca().spines['top'].set_visible(False)
    plt.gca().spines['right'].set_visible(False)
    plt.gca().spines['bottom'].set_visible(False)
    plt.gca().spines['left'].set_visible(False)
    plt.yticks([])
    plt.xticks(range(len(conditions)),conditions[new_order],rotation=30,ha='right')

    plt.subplot(gs[3])
    linkage(normed_data)
    dendrogram(feature_link,orientation='left');
    plt.axis('off');
```

```python
interact(plot_new_thing,width=(0,30,0.5),height=(0,30,0.5),height_ratio=(1,10,0.2));
```
