```python
def rename_func(x):
    "Return part of name prior to parentheses"
    if x.startswith("Cell"):
        return x.split('(')[0].lstrip("Cell: ").rstrip('/').rstrip(' ').lstrip(' ')
    else:
        return x
```

```python
def header_to_column(series):
    """Given a Series, return a Dataframe with columns: value, label.
    Created by associating header with each value in Series."""
    return pd.DataFrame(dict(values = series.values,
                             label = series.name))
```

```python
def headers_to_column(df):
    """Given Dataframe, return new Dataframe with two columns: value, label. 
    Created by taking each value in table, and pairing it with its column name."""
    return pd.concat([header_to_column(df[col]) for col in df.columns])
```

```python
def move_column(df,col,pos):
    """Return new Dataframe where col in moved to pos.
    if pos is 0, then first column becomes col, and all other columns are shifted right."""
    other_cols = [c for c in df.columns if c != col]
    return df[other_cols[0:pos] + [col] + other_cols[pos:]]
```

```python
def get_grid(n):
    "Return number of rows and columns for a given number of plots."
    rows = int(np.floor(np.sqrt(n)))
    cols = int(np.ceil(n/float(rows)))
    return rows,cols
```

```python
def plot_dist(ax,y):
    kde = stats.gaussian_kde(y)
    x = np.linspace(y.min(),y.max(),1000)
    y_fit = kde(x)
    rescale_factor = (np.histogram(y,bins=50)[0].max() / y_fit.max()) * 0.9

    ax.plot(x,y_fit*rescale_factor,linewidth = 2)

    plt.yticks([])
    ax.spines['left'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.get_xaxis().tick_bottom()
    
def chart_dists3(dists_list):
    """Create a chart for each distribution.
    dists_list: [[(string,[num])]]
    each element of dists_list has a subplot, and each subelement has a labelled histogram."""
    
    sns.set(style="white", palette="muted")
    b,g = sns.color_palette("muted",2)
    n_rows, n_cols = get_grid(len(dists_list))
    f,axes_grid = plt.subplots(n_rows,n_cols,figsize=(15,9))
    axes = np.array(axes_grid).flatten();
    
    for i, dists in enumerate(dists_list):
        for dist,color in zip(dists,sns.color_palette('muted',len(dists))):
            label = dist[0]
            plot_dist(axes[i],dist[1])
            axes[i].set_xlim(0,1)
            axes[i].set_ylim(0,250)
            if len(label) != 0:
                axes[i].set_xlabel(label,color=color)
    f.suptitle('Area_TxRed')
    plt.setp(axes, yticks=[],xticks=[])
```

```python
def chart_dists(dists_list):
    """Create a chart for each distribution.
    dists_list: [[(string,[num])]]
    each element of dists_list has a subplot, and each subelement has a labelled histogram."""
    
    sns.set(style="white", palette="muted")
    b,g = sns.color_palette("muted",2)
    n_rows, n_cols = get_grid(len(dists_list))
    f,axes_grid = plt.subplots(n_rows,n_cols,figsize=(15,9))
    axes = np.array(axes_grid).flatten();
    
    for i, dists in enumerate(dists_list):
        for dist,color in zip(dists,sns.color_palette('muted',len(dists))):
            sns.distplot(dist[1],color=color,ax=axes[i],hist=True,norm_hist=False)
            axes[i].set_xlim(0,0.5)
            axes[i].set_ylim(0,20)
            axes[i].text(0,-2,dist[0],color=color)
            
    sns.despine(left=True)
    plt.setp(axes, yticks=[],xticks=[])
    #plt.tight_layout()
```

```python
chart_dists(dists)
```

```python
def concatenate(x):
    "Concatenate a sublists into single list."
    if len(x) == 1:
        return x[0]
    elif len(x) == 2:
        return x[0] + x[1]
    else: 
        return x[0] + x[1] + concatenate(x[2:])
```

```python
def group_by_keyword(x,groups):
    return [[xi for xi in x if g in xi] for g in groups]
```

```python
def print_groups(groups,labels):
    for g,label in zip(groups,labels): 
        print "GROUP: {}".format(label)
        for gi in g: 
            print gi
        print ""    
```
